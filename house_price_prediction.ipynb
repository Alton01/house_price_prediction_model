```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('housing.csv')
```

```python
data
```

```python
# To check for null values
data.info()
```

```python
# To drop off null values and update data object
data.dropna(inplace=True)
```

```python
data.info()
```

```python
# split dataset into x and y representing independent variables and dependent variable
from sklearn.model_selection import train_test_split

x = data.drop(['median_house_value'], axis=1)
y = data['median_house_value']
```

```python
# split dataset into train and test splits

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
train_data = x_train.join(y_train)
```

```python
train_data
```

```python
# To get multiple histograms for the distribution of the individual features
train_data.hist(figsize=(10,10))
```

```python
#This code applies a logarithmic transformation to several columns
# The np.log() function calculates the natural logarithm of each element in the specified columns. Adding +1 inside the logarithm.
# Adding +1 inside the logarithm is a common practice to handle potential zero values in the original data.
# Logarithmic transformations are often used in data preprocessing to:
# Reduce skewness in the data, Stabilize the variance, Make relationships between variables more linear.
# This is helpful for improving performance of machine learning models that assume normally distributed data or linear relationships.
train_data['total_rooms'] = np.log(train_data['total_rooms'] + 1)
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms'] + 1)
train_data['population'] = np.log(train_data['population'] + 1)
train_data['households'] = np.log(train_data['households'] + 1)
```

```python
train_data.hist(figsize=(15,8))
```

```python
# implement One-hot encoding to ocean_proximity which is conversion of categorical information classes to a format where each class is a feature
# Here, we have implemented one hot-encoding, thus using binary 0 or 1 to encode which of the classes for ocean_proximity the property falls into.
# We have also updated the columns for the train data and dropped the ocean_proximity column
train_data = train_data.join(pd.get_dummies(train_data.ocean_proximity, dtype=int)).drop(['ocean_proximity'], axis=1)
```

```python
train_data
```

```python
# To see how the values of the table correlate with one another
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot=True)
```

```python
# to visualize the coordinates for median_house_value
# red signifies more expensive, blue signifies less expensive
plt.figure(figsize=(15,8))
sns.scatterplot(x='latitude', y='longitude', hue='median_house_value', data=train_data, palette='coolwarm')
```
