```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('housing.csv')
```

```python
data
```

```python
# To check for null values
data.info()
```

```python
# To drop off null values and update data object
data.dropna(inplace=True)
```

```python
data.info()
```

```python
# split dataset into x and y representing independent variables and dependent variable
from sklearn.model_selection import train_test_split

x = data.drop(['median_house_value'], axis=1)
y = data['median_house_value']
```

```python
# split dataset into train and test splits

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
train_data = x_train.join(y_train)
```

```python
train_data
```

```python
# To get multiple histograms for the distribution of the individual features
train_data.hist(figsize=(10,10))
```

```python
#This code applies a logarithmic transformation to several columns
# The np.log() function calculates the natural logarithm of each element in the specified columns. Adding +1 inside the logarithm.
# Adding +1 inside the logarithm is a common practice to handle potential zero values in the original data.
# Logarithmic transformations are often used in data preprocessing to:
# Reduce skewness in the data, Stabilize the variance, Make relationships between variables more linear.
# This is helpful for improving performance of machine learning models that assume normally distributed data or linear relationships.
train_data['total_rooms'] = np.log(train_data['total_rooms'] + 1)
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms'] + 1)
train_data['population'] = np.log(train_data['population'] + 1)
train_data['households'] = np.log(train_data['households'] + 1)
```

```python
train_data.hist(figsize=(15,8))
```

```python
# implement One-hot encoding to ocean_proximity which is conversion of categorical information classes to a format where each class is a feature
# Here, we have implemented one hot-encoding, thus using binary 0 or 1 to encode which of the classes for ocean_proximity the property falls into.
# We have also updated the columns for the train data and dropped the ocean_proximity column
train_data = train_data.join(pd.get_dummies(train_data.ocean_proximity, dtype=int)).drop(['ocean_proximity'], axis=1)
```

```python
train_data
```

```python
# To see how the values of the table correlate with one another
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot=True)
```

```python
# to visualize the coordinates for median_house_value
# red signifies more expensive, blue signifies less expensive
plt.figure(figsize=(15,8))
sns.scatterplot(x='latitude', y='longitude', hue='median_house_value', data=train_data, palette='coolwarm')
```

```python
# 1) to calculate ratio of rooms that are bedrooms
# 2) to calculate how many rooms per household

train_data['bedroom_ratio'] = train_data['total_bedrooms'] / train_data['total_rooms']
train_data['household_rooms'] = train_data['total_rooms'] / train_data['households']
```

```python
# Performing simple linear regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler

# split data into x and y variables
x_train, y_train = train_data.drop(['median_house_value'], axis=1), train_data['median_house_value']

reg = LinearRegression()

# allows model to learn the relationship between the features and the target variable from the training data.
reg.fit(x_train, y_train)
```

```python
# Preprocess Test Data
test_data = x_test.join(y_test)

test_data['total_rooms'] = np.log(test_data['total_rooms'] + 1)
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms'] + 1)
test_data['population'] = np.log(test_data['population'] + 1)
test_data['households'] = np.log(test_data['households'] + 1)

test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity, dtype=int)).drop(['ocean_proximity'], axis=1)

test_data['bedroom_ratio'] = test_data['total_bedrooms'] / test_data['total_rooms']
test_data['household_rooms'] = test_data['total_rooms'] / test_data['households']

# split data into x and y variables
x_test, y_test = test_data.drop(['median_house_value'], axis=1), test_data['median_house_value']
```

```python
test_data
```

```python
# evaluate performance of the trained model by comparing its predictions on the test_data.
reg.score(x_test, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
#trains the RandomForestRegressor model using the fit()
forest = RandomForestRegressor()

# allows the Random Forest model to learn the complex relationships between the features and the target variable
forest.fit(x_train, y_train)
```

```python
forest.score(x_test, y_test)
```

```python
# performs a grid search to find the best hyperparameters for the RandomForestRegressor model.
# GridSearchCV is a technique used to systematically work through multiple combinations of hyperparameter values.
# cross-validating each combination to find the best performing set of hyperparameters.

from sklearn.model_selection import GridSearchCV

forest = RandomForestRegressor()

param_grid = {
    "n_estimators": [30, 70, 100],
    "max_features": [2, 4, 6, 8]
}

# cv=5: Specifies that 5-fold cross-validation should be used. Training data will be split into 5 folds,
grid_search = GridSearchCV(forest, param_grid, cv=5, scoring='neg_mean_squared_error', return_train_score=True)

grid_search.fit(x_train, y_train)
```

```python
grid_search.best_estimator_
```

```python
best_forest = grid_search.best_estimator_
```

```python
best_forest.score(x_test, y_test)
```
