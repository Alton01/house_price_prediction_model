```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('housing.csv')
```

```python
data
```

```python
# To check for null values
data.info()
```

```python
# To drop off null values and update data object
data.dropna(inplace=True)
```

```python
data.info()
```

```python
# split dataset into x and y representing independent variables and dependent variable
from sklearn.model_selection import train_test_split

x = data.drop(['median_house_value'], axis=1)
y = data['median_house_value']
```

```python
# split dataset into train and test splits

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
train_data = x_train.join(y_train)
```

```python
train_data
```

```python
train_data = train_data.drop(['ocean_proximity'], axis=1)
```

```python
# To get multiple histograms for the distribution of the individual features
train_data.hist(figsize=(10,10))
```

```python
# To get correlation value between variables
train_data.corr()
```

```python
# to produce a correlation heatmap to view correlation values between variables
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot=True)
```
